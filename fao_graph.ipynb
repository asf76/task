# creating few Plot using seaborn and and matplotlib

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

foe = pd.read_csv("FAOSTAT_data_en_10-7-2022.csv")
```

```python
foe.head(5)
```

```python

# draw catplot 
sns.catplot(x="Domain", y="Year",hue="Item", data=foe) 
plt.title("FOA DATA")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(hue='Domain', x="Year", y="Value", data=foe)
plt.title("FOA DATA By Value and Year")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.set_color_codes("pastel")
#sns.barplot(x="Domain", y="Year", data=foe,
#            hue='Item', color="b", order=["Year", "Value"])

sns.barplot(hue='Domain', x="Item", y="Value",  data=foe)
plt.title("FOA DATA By Item and Value")
plt.show()
```

```python
sns.boxplot(x="Year", y="Item", data=foe, saturation=0.6)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x="Year", y="Element", hue= "Item", data=foe, order=["Year", "Value"])
#, color="red", 
#            ci=None, palette='pastel') 
plt.show()
```
